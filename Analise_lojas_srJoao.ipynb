### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head(-1)
```

#1. Análise do faturamento


```python
import matplotlib.pyplot as plt
import numpy as np

# Remove espaços em branco no início e no final dos nomes das colunas.
for df in [loja1, loja2, loja3, loja4]:
    df.columns = df.columns.str.strip()

# Faturamento total de cada loja
fat_loja1 = round(sum(loja1['Preço']), 2)
fat_loja2 = round(sum(loja2['Preço']), 2)
fat_loja3 = round(sum(loja3['Preço']), 2)
fat_loja4 = round(sum(loja4['Preço']), 2)
fat_geral = fat_loja1+fat_loja2+fat_loja3+fat_loja4

# Lista do faturamento total de cada loja
fat_lojas = [fat_loja1, fat_loja2, fat_loja3, fat_loja4]

# Construção do gráfico

titulo_principal = "Faturamento por loja"
titulo_eixo_x = "Lojas"
titulo_eixo_y = "FATURAMENTO (R$)"
graf_lojas = ["loja1", "loja2", "loja3", "loja4"]
cores = ["#696969", "#808080", "#A9A9A9", "#C0C0C0"]


plt.rcParams.update({"font.size":14}) #define o padrao de tamanho das fontes dos gráficos
plt.figure(figsize=(10,6))
plt.title(titulo_principal)
plt.xlabel(titulo_eixo_x) #define o título do gráfico no eixo x (<titulo>, <fontsize = ?>, <color = "cor">, <edgecolor = "cor">, <linewidth = ?> )
plt.ylabel(titulo_eixo_y) #define o título do gráfico no eixo y
plt.bar(graf_lojas, fat_lojas, color = cores) #cria o gráfico de barras
plt.figtext(0.5, -0.05, f"Total Geral de Faturamento: R$ {fat_geral:,.2f}", #insere um texto na figura
            ha="center", fontsize=12, weight='bold', color='black')
for i, valor in enumerate(fat_lojas):
    plt.text(i, valor + 10000, f"R$ {valor:,.0f}", ha='center', va='bottom', fontsize=10) #adiciona um texto dentro da area do gráfico. <x, y, texto>
plt.savefig("analise_faturamento.png", dpi = 600) #salvar a img do gráfico
plt.show()



















```

# 2. Vendas por Categoria


```python
print("Loja 1 - Produtos por Categoria")
print(loja1["Categoria do Produto"].value_counts()) #retorna Uma Series contendo as contagens, ordenada por frequência decrescente.
print("\n")

print("Loja 2 - Produtos por Categoria")
print(loja2["Categoria do Produto"].value_counts())
print("\n")

print("Loja 3 - Produtos por Categoria")
print(loja3["Categoria do Produto"].value_counts())
print("\n")

print("Loja 4 - Produtos por Categoria")
print(loja4["Categoria do Produto"].value_counts())
print("\n")


```

```python


loja1_produtos_por_categoria = loja1["Categoria do Produto"].value_counts()
loja2_produtos_por_categoria = loja2["Categoria do Produto"].value_counts()
loja3_produtos_por_categoria = loja3["Categoria do Produto"].value_counts()
loja4_produtos_por_categoria = loja4["Categoria do Produto"].value_counts()

df_categorias = pd.DataFrame({"loja 1": loja1_produtos_por_categoria,
                              "loja 2": loja2_produtos_por_categoria,
                              "loja 3": loja3_produtos_por_categoria,
                              "loja 4": loja4_produtos_por_categoria}).fillna(0)

cores = ['#FF0000', '#FFA500', '#00BFFF', '#EE82EE']
df_categorias.plot(kind="bar", figsize=(12, 6), color=cores) #Cria um gráfico de barras
plt.title("Quantidade de Produtos por Categoria")
plt.ylabel("Quantidade")
plt.xticks(rotation=45) #Rotaciona os rótulos do eixo X em 45 graus
plt.legend(title="Lojas") #Mostra uma legenda no gráfico
plt.tight_layout() #Ajusta automaticamente o layout do gráfico
plt.savefig("vendas_categoria.png", dpi = 600)
plt.show()




```

# 3. Média de Avaliação das Lojas

```python
# Calcula a média de avaliação para cada loja
media_avaliacao_loja1 = loja1["Avaliação da compra"].mean()
media_avaliacao_loja2 = loja2["Avaliação da compra"].mean()
media_avaliacao_loja3 = loja3["Avaliação da compra"].mean()
media_avaliacao_loja4 = loja4["Avaliação da compra"].mean()

# Arredonda as médias para 2 casas decimais para exibição
media_avaliacao_loja1_arredondada = round(media_avaliacao_loja1, 2)
media_avaliacao_loja2_arredondada = round(media_avaliacao_loja2, 2)
media_avaliacao_loja3_arredondada = round(media_avaliacao_loja3, 2)
media_avaliacao_loja4_arredondada = round(media_avaliacao_loja4, 2)


cores = ['#FF0000', '#FFA500', '#00BFFF', '#EE82EE']
medias_ava_lojas = [media_avaliacao_loja1_arredondada, media_avaliacao_loja2_arredondada, media_avaliacao_loja3_arredondada, media_avaliacao_loja4_arredondada]
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]

# Cria a figura *antes* de adicionar os elementos do gráfico
plt.figure(figsize=(10, 6))

# Cria o gráfico de dispersão
plt.scatter(lojas, medias_ava_lojas, color=cores, s=150)
plt.ylabel("Nota média")
plt.title("Média de avaliação das lojas")
plt.ylim(3.8, 4.2)
plt.grid(True, linestyle='--', alpha=0.5)

# Adiciona os valores das notas no gráfico
for i, nota in enumerate(medias_ava_lojas):
    plt.text(i, nota + 0.02, f"{nota:.2f}", ha='center', va='bottom', fontsize=10)

# Adiciona a linha conectando os pontos
plt.plot(lojas, medias_ava_lojas, color='gray', linestyle='dashed', linewidth=2, marker='o')

# Salva a figura *depois* de adicionar os elementos do gráfico
plt.savefig("media_avaliacao.png", dpi=600)
plt.show()




```

# 4. Produtos Mais e Menos Vendidos

```python
#Quantidade de vendas de cada produto por loja, de forma decrescente
quant_produto_loja1 = loja1.groupby("Produto").size().sort_values(ascending=False)
quant_produto_loja2 = loja2.groupby("Produto").size().sort_values(ascending=False)
quant_produto_loja3 = loja3.groupby("Produto").size().sort_values(ascending=False)
quant_produto_loja4 = loja4.groupby("Produto").size().sort_values(ascending=False)

#Dicionário com os produtos
dic_prod_lojas = {
    "loja 1" : quant_produto_loja1,
    "loja 2" : quant_produto_loja2,
    "loja 3" : quant_produto_loja3,
    "loja 4" : quant_produto_loja4
}

#Encontrar o produto mais e menos vendido
todos_produtos = pd.concat([quant_produto_loja1, quant_produto_loja2, quant_produto_loja3, quant_produto_loja4])
prod_mais_vendido = todos_produtos.groupby(todos_produtos.index).sum().idxmax()
prod_menos_vendido = todos_produtos.groupby(todos_produtos.index).sum().idxmin()
quant_mais_vendida = todos_produtos.groupby(todos_produtos.index).sum().max()
quant_menos_vendida = todos_produtos.groupby(todos_produtos.index).sum().min()

#lista para armazenar os produtos e quantidades por loja
produtos_mais = []
quantidades_mais = []
produtos_menos = []
quantidades_menos = []

#Preencher as listas com os dados do dicionário
for nome_loja, dados in dic_prod_lojas.items():
    produtos_mais.append(dados.idxmax())
    quantidades_mais.append(dados.max())
    produtos_menos.append(dados.idxmin())
    quantidades_menos.append(dados.min())

#Gráfico
x = np.arange(len(lojas))  # posição das lojas no eixo x
largura = 0.35
plt.figure(figsize=(12, 7))
plt.style.use('ggplot')
# Barras para produtos mais vendidos
barras_mais = plt.bar(x - largura/2, quantidades_mais, width=largura, label='Mais Vendido', color='#00BFFF')

# Barras para produtos menos vendidos
barras_menos = plt.bar(x + largura/2, quantidades_menos, width=largura, label='Menos Vendido', color='#FF0000')

# Adiciona os nomes dos produtos nas barras
for i in range(len(lojas)):
    plt.text(x[i] - largura/2, quantidades_mais[i] + 0.1, produtos_mais[i], ha='center', va='bottom', fontsize=9, rotation=0)
    plt.text(x[i] + largura/2, quantidades_menos[i] + 0.1, produtos_menos[i], ha='center', va='bottom', fontsize=9, rotation=0)

# Eixos e legendas
plt.xticks(x, lojas)
plt.ylabel('Quantidade Vendida')
plt.title('Produto Mais e Menos Vendido por Loja', fontweight='bold')
plt.legend()
plt.tight_layout()

# Adicionando anotação explicativa
plt.figtext(0.5, 0.01,
            f"O produto mais vendido no geral é '{prod_mais_vendido}' com {quant_mais_vendida} unidades vendidas.\n O produto menos vendido no geral é '{prod_menos_vendido}' com {quant_menos_vendida} unidades vendidas.",
            ha='center', fontsize=12)

plt.savefig("mais_menos_vend.png", dpi = 600)
plt.subplots_adjust(bottom=0.1)  # Aumenta o espaço na parte inferior para o texto
plt.ylim(20,)
plt.show()








```

# 5. Frete Médio por Loja


```python
import matplotlib.ticker as mtick

lista_lojas = [loja1, loja2, loja3, loja4]

# Cria listas para armazenar os valores
media_fretes = []
total_fretes = []

# Calcula os valores para cada loja
for df in lista_lojas:
    media_fretes.append(df['Frete'].mean())
    total_fretes.append(df['Frete'].sum())

df_frete = pd.DataFrame({
    'Loja': lojas,
    'Média Frete': media_fretes,
    'Total Frete': total_fretes
})

plt.figure(figsize=(10, 6))
plt.scatter(df_frete['Total Frete'], df_frete['Média Frete'], color=['#FF0000', '#FFA500', '#00BFFF', '#EE82EE'], s=150)

def format_reais(x, pos):
  """Formata o valor x como moeda brasileira (R$)."""
  return f'R$ {x:,.0f}'



for i in range(len(df_frete)):
    plt.text(df_frete['Total Frete'][i],
             df_frete['Média Frete'][i] +.2,
             df_frete['Loja'][i],
             fontsize=12,
             ha='right',
             va='center')



plt.xlabel('Total de Frete (R$)')
plt.ylabel('Média de Frete (R$)')
plt.title('Dispersão: Média x Total de Frete por Loja')
plt.grid(True)
plt.tight_layout()
plt.ylim(31, 35.5)
plt.gca().xaxis.set_major_formatter(format_reais)
plt.xlim(72000,82000)
plt.savefig("frete_medio", dpi = 600)
plt.show()


```
